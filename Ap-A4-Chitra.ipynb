```python
#finding unique words 
from collections import Counter
Count = Counter(split_data)
Unique_count = Count.most_common(1)
print('Most unique word is ', Unique_count)
```

```python
most_appearing_word = 0
for i in split_data:
    if i.startswith('t'):
        most_appearing_word +=1
print('the words count starting with t is ', most_appearing_word)
```

```python
df = pd.read_csv('/Users/chitra/Documents/Subjects/Analytic Python/assignment/breast_cancer-1.csv')
```

```python
df.head(10)
```

```python
#dropping rows which has null
df.dropna(axis=0, inplace=True)
```

```python
df.isnull().sum()
```

```python
# making class label ‘benign’ as 0 and ‘malignant’ as 1 in the dataset using dictionary
df['class'].replace({'benign': 0,'malignant':1}, inplace=True)
```

```python
# checking the class values
print(df['class'].unique())
```

```python
# creating new variable cell_type_label and setting initial value as null
df['cell_type_label'] = np.NaN
```

```python
#checking newly created columns 
df
```

```python
#based on condition filing the value for cell_type_label
for i in range(df.shape[0]):
    if df.iloc[i]['cell_size_uniformity'] > 5.0 and df.iloc[i]['cell_shape_uniformity'] > 6.0:
        df['cell_type_label'].iloc[i] = 'Normal'    
    else:
        df['cell_type_label'].iloc[i] = 'Abnormal'
```
