# 1.	Use the randn function to create an array with a dimension of 5X5, and use a for loop to calculate the sum of all elements in the diagonal of the array

```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
array = np.random.randn(5, 5)
print("2D Array filled with random values : \n", array);
```

```python
length = len(array)
r1 = 0
r2 = 0
for i in range(length):
    r1 += array[i][length - i - 1] #first diagnol
    r2 += array[i][i] # second opposite diagnol
print(r1 + r2)
# sum  absolute values
print(abs(r1) + abs(r2))
```

# 2. 2	Randomly generate a 8x9 array from a normal distribute with mean = 1, sigma = 0.5. Calculate the mean of elements whose indexes have a relation of (i+j)%5 == 0  (i is row index and j is column index).	(10)

```python
x = np.random.normal(loc=1, scale=0.5, size = (8, 9))
print(x)
add = 0
count = 0
for i in range(8):
    for j in range(9):
        if i+j%5 == 0:
            add = add + x[i][j]
            count += 1
print(f'The mean is:{add/count}')
```

```python
col = ['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am',
       'gear', 'carb']
def d_plots(col):
    for i in col:
        sns.distplot(df[i])
        plt.axvline(df[i].values.mean(), color='yellow', linestyle='--', linewidth=1)
        print(f'Mean for {i} is {df[i].mean()}')
        print(f'Median for {i} is {df[i].median()}')
        print(f'Standard Deviation for {i} is {df[i].std()}\n')  
        plt.show()
          
d_plots(col)
```
