### Train and validate data (Machine learning)


```python
#dataset spliting   
from sklearn import *
from sklearn.model_selection import train_test_split
array = df.values    
X = array[:,0:4]    
Y = array[:,4]    
validation_size = 0.20    
seed = 7    
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=validation_size,     
random_state=seed)    
 #k=10    
num_folds = 10    
num_instances = len(X_train)    
seed = 7    
scoring = 'accuracy' 
```

### we are going to evaluate five famous frequently used algorithms such as
    Logistic regression
    Decision tree classifier
    Support Vector Machine

```python
#evaluate model to determine better algorithm 
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.linear_model import *
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

models = []    
models.append(('Linear Regression', LogisticRegression()))       
models.append(('CART', DecisionTreeClassifier()))    
models.append(('SVM', SVC()))    
    
results = []    
names = []    
for name, model in models:    
    kfold = model_selection.KFold(n_splits=10, random_state=None, shuffle=True ) #n=num_instances, n_folds=num_folds, random_state=seed)    
    cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)    
    results.append(cv_results)    
    names.append(name)    
    msg = "%s: %f" % (name, cv_results.mean())    
    print(msg)   
```

### SVM is doing best compared to other model with 97% of accuracy

### Validate the data(prediction)

```python
#prediction 
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
svn = SVC()    
svn.fit(X_train, Y_train)    
predictions = svn.predict(X_validation)    
print(accuracy_score(Y_validation, predictions))    
print(confusion_matrix(Y_validation, predictions))    
print(classification_report(Y_validation, predictions))
```

### accuracy score, confusion matrix and classification reportsa are shown above.
