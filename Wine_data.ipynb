```python
# checking the count of these 
# dropping the outliers using IQR
def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers

#df[find_outliers_IQR(df).notnull()

find_outliers_IQR(df).count()
```

### Splitting the data into Test and Train/Modelling

```python
# dataset spliting   
from sklearn import *
from sklearn.model_selection import train_test_split
array = df.values    
X = array[:,0:13]    
Y = array[:,13]    
validation_size = 0.20    
seed = 7    
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X, Y, test_size=validation_size,     
random_state=seed)    
#k=10    
num_folds = 10    
num_instances = len(X_train)    
seed = 7    
scoring = 'accuracy' 
```

Here spitting the data into Test and validation

The validation size taken is 20.
So the Data is split into 80 : 20 fashion

Also I will be Validating using K -cross validation with K=10

### Evaluating of Data 

```python
# Fitting the model here
name = 'Decision Tree'
model = DecisionTreeClassifier()
kfold = model_selection.KFold(n_splits=10, random_state=None, shuffle=True ) #n=num_instances, n_folds=num_folds, random_state=seed)    
cv_results = model_selection.cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)    
results = cv_results       
msg = "%s: %f" % (name, cv_results.mean())    
print(msg)   
```

As you can see, I'm using Decision Tree model, as we have no normal distribution of data, decision tree would perform better in such cases. 

In the test set, the performed with accuracy 88% with cross validation fold = 10

### Validating the data(prediction)

```python
#prediction 
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
Dt = DecisionTreeClassifier()    
Dt.fit(X_train, Y_train)    
predictions = Dt.predict(X_validation) 
print('Accuracy')
print(accuracy_score(Y_validation, predictions)) 
print('--------------------------------------------------------------------------------')
print('confusion matrix')
print(confusion_matrix(Y_validation, predictions))   
print('--------------------------------------------------------------------------------')
print('classification report')
print(classification_report(Y_validation, predictions))
```

So in the above step I have validated the train data. 

As we can see the primary result metrics taken here is Accuracy

confusion matrix talks about performance of our classification model

Also the classification report which gives precision, recall, f1-score for each class.

Here I have shown classification report. 

1. Precision : Percentage of correct positive predictions relative to total positive predictions.
2. Recall: Percentage of correct positive predictions relative to total actual positives.
3. F1 Score: A weighted harmonic mean of precision and recall. The closer to 1, the better the model.
4. Support: These values simply tell us how many data point belonged to each class in the test dataset. 

We can see that accuracy for all 3 classes to be 92%
